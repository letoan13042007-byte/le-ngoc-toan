Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def kiem_tra_so_amstrong(x):
    s = str(x)
    n = len(s)
    tong = 0
    for i in s:
        tong += int(i) ** n
    
    if tong == x :
        return True 
    else :
        return False
kiem_tra_so_amstrong(111)


       

```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
detect_number_type(10)
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    a1 = x2 - x1
    b1 = y2 - y1
    a2 = x3 - x1
    b2 = y3 - y2
    if a1 / a2 == b1 / b2 :
        return True
    else:
        return False
are_collinear(1,2,3,4,5,6)
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    xp, yp = p
    list = []
    for i in range(4):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % 4]
        d = (x2 - x1) * (yp - y1) - (y2 - y1) * (xp - x1)
        if d > 0: list.append(1)
        elif d < 0: list.append(-1)
    return len(set(list)) <= 1
quad = [(0,0), (4,0), (4,3), (0,3)]
p = (2,1)
print(point_in_quadrilateral(quad, p))
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count(numbers):
    total_count = 0
    for i in numbers:
        s = str(abs(i))
        for char in s:
            if char != '0':
                total_count += 1
    return total_count
lst = [100,210,230]
print(count(lst))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def tinh_goc_abc(x1, y1, x2, y2, x3, y3):
    a2 = (x3 - x2)**2 + (y3 - y2)**2
    c2 = (x1 - x2)**2 + (y1 - y2)**2
    b2 = (x3 - x1)**2 + (y3 - y1)**2
    a, c = math.sqrt(a2), math.sqrt(c2)
    if a == 0 or c == 0:
        return 0.0
    cos_b = (a2 + c2 - b2) / (2 * a * c)
    cos_b = max(-1.0, min(1.0, cos_b))
    return math.degrees(math.acos(cos_b))

print(tinh_goc_abc(1,2,3,4,5,6))
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4 :
        if 0 < km <= 0.8 :
            money = km * 11000
        elif 0.8 < km <= 20 :
            money = 0.8 * 11000 + (km - 0.8) * 13000
        elif km > 20 :   
            money = 0.8 * 11000 + 19.2 * 13000 + (km - 20) * 10500
    elif car_type == 7 :
        if 0 < km <= 0.8 :
            money = km * 12000
        elif 0.8 < km <= 20 :
            money = 0.8 * 11000 + (km - 0.8) * 14000
        elif km > 20 :   
            money = 0.8 * 11000 + 19.2 * 14000 + (km - 20) * 12000
    if wait_minutes > 5 :
        phicho = (round(wait_minutes / 5 ) - 1) * 2000
    if start_hour in [22,23,0,1,2,3,4,5] :
        total =  1.2 * (money + phicho)
    else :
        total = money + phicho
    if km > 30 and not start_hour in [22,23,0,1,2,3,4,5]:
        total = 0.9 * total
    if total < 50000 and km > 5 :
        total = 50000
    return int(total)
smart_taxi_fare(4, 7, 60, 10)
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def check(x):
    s = str(x)
    return len(s) == len(set(s))
def sum_1_to_n(n):
    if n <= 0:
        return 0
    return sum(x for x in range(1, n + 1) if check(x))
print(sum_1_to_n(12))
```
