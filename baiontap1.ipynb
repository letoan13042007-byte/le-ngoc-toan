1 Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.start_hour thuộc [22,23,0,1,2,3,4,5]
4. Phụ thu ban đêm: nếu , cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4 :
        if 0 < km <= 0.8 :
            money = km * 11000
        elif 0.8 < km <= 20 :
            money = 0.8 * 11000 + (km - 0.8) * 13000
        elif km > 20 :   
            money = 0.8 * 11000 + 19.2 * 13000 + (km - 20) * 10500
    elif car_type == 7 :
        if 0 < km <= 0.8 :
            money = km * 12000
        elif 0.8 < km <= 20 :
            money = 0.8 * 11000 + (km - 0.8) * 14000
        elif km > 20 :   
            money = 0.8 * 11000 + 19.2 * 14000 + (km - 20) * 12000
    if wait_minutes > 5 :
        phicho = (round(wait_minutes / 5 ) - 1) * 2000
    if start_hour in [22,23,0,1,2,3,4,5] :
        total =  1.2 * (money + phicho)
    else :
        total = money + phicho
    if km > 30 and not start_hour in [22,23,0,1,2,3,4,5]:
        total = 0.9 * total
    if total < 50000 and km > 5 :
        total = 50000
    return int(total)
smart_taxi_fare(4, 7, 60, 10)

```

5 Yêu cầu: Cho ba số a, b, c, trả về chúng dưới dạng danh sách theo thứ tự không giảm. 

```python
def sap_xep(a, b, c):
    if a > b:
        a, b = b, a
    if a > c:
        a, c = c, a
    if b > c:
        b, c = c, b    
    return [a, b, c]
print(sap_xep(2,1,3))  
```

6 Yêu cầu: Cho năm số a, b, c, d, e, trả về True nếu tồn tại bất kỳ ba số nào trong đó bằng nhau; ngược lại trả về False. 
Không dùng tuple hay dictionary. 

```python
def kiem_tra(a, b, c, d, e):
    list = [a, b, c, d, e]
    for so in list:
        if list.count(so) >= 3:
            return True
    return False
print(kiem_tra(1, 2, 3, 4, 5))  
```

7 Yêu cầu: Cho ba số a, b, c, trả về giá trị ở giữa khi sắp xếp ba số theo thứ tự tăng dần (giá trị lớn thứ hai). 
Định nghĩa này vẫn đúng khi có phần tử trùng nhau (ví dụ: 2, 2, 5 → giá trị giữa là 2). 

```python
def tim_so_o_giua(a, b, c):
    if (b <= a <= c) or (c <= a <= b):
        return a
    if (a <= b <= c) or (c <= b <= a):
        return b
    return c
print(tim_so_o_giua(1,2,3))
```

8 Yêu cầu: Cho hai ngày sinh dưới dạng (day, month, year) của người 1 và người 2, trả về 1 nếu người 1 lớn tuổi hơn (ngày 
sinh sớm hơn), 2 nếu người 2 lớn tuổi hơn, và 0 nếu bằng nhau. 
Ví dụ: 27 2 2007 so với 28 2 2007 → trả về 1 vì ngày 27 sớm hơn nên người 1 lớn tuổi hơn.

```python
def so_sanh_tuoi(d1, m1, y1, d2, m2, y2):
    if y1 < y2:
        return 1
    elif y1 > y2:
        return 2
    if m1 < m2:
        return 1
    elif m1 > m2:
        return 2
    if d1 < d2:
        return 1
    elif d1 > d2:
        return 2
    return 0
print(so_sanh_tuoi(13,4,2007,1,1,2007))  
```

9 Yêu cầu: Cho số nguyên n, trả về: - 5n + 1 nếu n mod 3 == 1 - 6n + 2 nếu n mod 3 == 2 - 7n + 3 nếu n mod 3 == 0 

```python
def tinh_gia_tri(n):
    du = n % 3   
    if du == 1:
        return 5 * n + 1
    elif du == 2:
        return 6 * n + 2
    else:  
        return 7 * n + 3
print(tinh_gia_tri(5))
```

16 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def tinh_goc_abc(x1, y1, x2, y2, x3, y3):
    a2 = (x3 - x2)**2 + (y3 - y2)**2
    c2 = (x1 - x2)**2 + (y1 - y2)**2
    b2 = (x3 - x1)**2 + (y3 - y1)**2
    a, c = math.sqrt(a2), math.sqrt(c2)
    if a == 0 or c == 0:
        return 0.0
    cos_b = (a2 + c2 - b2) / (2 * a * c)
    cos_b = max(-1.0, min(1.0, cos_b))
    return math.degrees(math.acos(cos_b))

print(tinh_goc_abc(1,2,3,4,5,6))
```

26 Yêu cầu: Viết hàm are_collinear(x1, y1, x2, y2, x3, y3) trả về True nếu ba điểm thẳng hàng, ngược lại trả về False. - Đầu vào: ba điểm x1, y1, x2, y2, x3, y3 biểu diễn tọa độ của ba điểm. - Đầu ra: giá trị boolean. 

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    a1 = x2 - x1
    b1 = y2 - y1
    a2 = x3 - x1
    b2 = y3 - y2
    if a1 / a2 == b1 / b2 :
        return True
    else:
        return False
are_collinear(1,2,3,4,5,6)
```

28 Yêu cầu: Viết hàm sum_halves(s) nhận chuỗi s gồm đúng 6 chữ số và trả về tổng của số nguyên 3 chữ số đầu và số nguyên 
3 chữ số cuối. - Đầu vào: chuỗi s độ dài 6, chỉ gồm các chữ số 0-9. - Đầu ra: tổng kiểu số nguyên. - Ví dụ: '100200' -> 100 + 200 = 300. - Nếu đầu vào không hợp lệ (sai độ dài hoặc chứa ký tự không phải chữ số), phát sinh ValueError. 

```python
def sum_halves(s):
    x = str(s)
    n = len(x)
    tongdau = 0
    tongcuoi = 0
    
    for i in range(n):
        if i <= 2 :
            tongdau += int(x[i])
            continue
        elif 3 <= i < 6 :
            tongcuoi += int(x[i])
    return tongcuoi + tongdau
sum_halves(123456)
    
```

29 Yêu cầu: Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại; ngược lại 
trả về False. - Đầu vào: các số a, b, c (int hoặc float). - Đầu ra: giá trị boolean. - Cần xét các trường hợp đặc biệt như số 0 và số âm.

```python
def is_one_product_of_others(a, b, c):
    return a == b * c or b == a * c or c == a * b
print(is_one_product_of_others(1,2,3))
```

30 Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu 
chấm động, ngược lại trả về 'neither'. - Đầu vào: chuỗi s. - Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. - Quy tắc: - Cho phép có dấu + hoặc - ở đầu. - Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). - Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). - Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
detect_number_type(10)
```
