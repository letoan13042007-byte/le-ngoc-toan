"[SumPrimesInRange] 
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)." 

```python
a , b = map(int, input().split())
demso = 0
for i in range(a, b + 1):
    if i  > 3  and i % 2 != 0 and i % 3 != 0  and i % 1 == 0:
        demso += 1
    elif i == 2 or i == 3 :
        demso += 1
    elif i < 2 :
        continue       
print(demso)     

                
```

"[LargestPrimeFactor] 
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 

```python
n = int(input())
list = []
if n >= 2 :
    for i in range(1,n + 1):
        if n % i == 0 :
            if i < 2 :
                continue
            elif i == 2 or i == 3:
                list.append(i)
            elif i > 3 and i % 2 != 0 and i % 3 != 0 and i % 1 == 0 :
                list.append(i) 
        elif  n % i != 0 :
            continue
    print(max(list))        

```

[EvenDivisors]
Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6

```python
n = abs(int(input()))
dem = 0
if n % 2 == 0:
    for i in range(2, n + 1, 2):
        if n % i == 0:
            dem += 1

print(dem)       
```
