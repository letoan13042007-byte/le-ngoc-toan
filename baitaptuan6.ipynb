Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ
nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
def check(danh_sach):
    result = []
    for i in danh_sach:
        if i not in result :
            result.append(i)
    return result
danh_sach = list(map(int, input().split()))
print(check(danh_sach))
```

W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó

```python
def tao_tu_dien_danh_sach():
    try:
        dong_vao = input().split()
        if not dong_vao:
            return     
        tu_dien = {}
        for cap in dong_vao:
            khoa, gia_tri = cap.split(':')
            if khoa not in tu_dien:
                tu_dien[khoa] = []
            tu_dien[khoa].append(gia_tri)      
        print(tu_dien)
    except (EOFError, ValueError):
        pass
print(tao_tu_dien_danh_sach())
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
key. 

```python
def nhap():
    d = {}
    for _ in range(int(input())):
        k = input()
        v = int(input())
        d[k] = v
    return d
d1 = nhap()
d2 = nhap()
for k, v in d2.items():
    d1[k] = d1.get(k, 0) + v

print(dict(sorted(d1.items())))
```

W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple

```python
def tim_cap_tong_bang_k():
    try:
        chuoi_vao = input().split()
        if not chuoi_vao:
            return
        cac_so = [int(x) for x in chuoi_vao]
        k = int(input())  
        ket_qua = []
        n = len(cac_so)
        for i in range(n):
            for j in range(i + 1, n):
                if cac_so[i] + cac_so[j] == k:
                    ket_qua.append((cac_so[i], cac_so[j]))
        ket_qua.sort(key=lambda x: x[0])
        print(ket_qua) 
    except (EOFError, ValueError):
        pass
print(tim_cap_tong_bang_k())
```

W6A12
[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
trong số đó. 

```python
danh_sach = list(map(int, input().split()))
dem = {}
for x in danh_sach:
    if x in dem:
        dem[x] += 1
    else:
        dem[x] = 1
max_count = 0
res = None
for k, v in dem.items():
    if v > max_count:
        max_count = v
        res = k
    elif v == max_count:
        if res is None or k < res:
            res = k
print(res)
```

[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một
dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In
ra dictionary mới.

```python
n = int(input())
my_dic = {}
for i in range(n):
    key = input()
    val = input()
    my_dic[key] = val
new_dic = {}
for i,j in my_dic.items():
    new_dic[j] = i
print(new_dic)

```

[Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp. 

```python
dong1 = list(map(int, input().split()))
dong2 = list(map(int, input().split()))
chung = []
for i in set(dong1) :
    for j in set(dong2):
        if i == j :
            chung.append(i)
print(chung)
```

[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
my_dic = {}
k = int(input())
n = int(input())
for i in range(n):
    key = input()
    val = int(input())
    my_dic[key] = val
new_dic = {}
for i in my_dic:
    if my_dic[i] > k :
        new_dic[i] = my_dic[i]
    else :
        continue
print(new_dic)

```
