W6A1
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ
nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
def loai_bo_trung_lap():
    try:
        chuoi_vao = input().split()
        if not chuoi_vao:
            return 
        cac_so = [int(x) for x in chuoi_vao]
        da_thay = set()
        ket_qua = []       
        for so in cac_so:
            if so not in da_thay:
                ket_qua.append(so)
                da_thay.add(so)     
        print(ket_qua)
    except EOFError:
        pass
loai_bo_trung_lap()
```

 W6A2
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị
trí i), giữ nguyên thứ tự

```python
def tong_tich_luy():
    try:
        chuoi_vao = input().split()
        if not chuoi_vao:
            return      
        cac_so = [int(x) for x in chuoi_vao]
        ket_qua = []
        tong_hien_tai = 0     
        for so in cac_so:
            tong_hien_tai += so
            ket_qua.append(tong_hien_tai)   
        print(ket_qua)
    except EOFError:
        pass
tong_tich_luy()
```

W6A3
Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)

```python
def xoay_trai_tuple():
    try:
        dong_dau = input().split()
        if not dong_dau:
            return
        
        bo_so = tuple(int(x) for x in dong_dau)
        k = int(input())
        
        n = len(bo_so)
        k = k % n
        
        ket_qua = bo_so[k:] + bo_so[:k]
        print(ket_qua)
    except (EOFError, ValueError, ZeroDivisionError):
        pass
xoay_trai_tuple()
```

W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó

```python
def tao_tu_dien_danh_sach():
    try:
        dong_vao = input().split()
        if not dong_vao:
            return     
        tu_dien = {}
        for cap in dong_vao:
            khoa, gia_tri = cap.split(':')
            if khoa not in tu_dien:
                tu_dien[khoa] = []
            tu_dien[khoa].append(gia_tri)      
        print(tu_dien)
    except (EOFError, ValueError):
        pass
print(tao_tu_dien_danh_sach())
```

W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
...} (đếm số >0, <0, =0)

```python
def dem_loai_so():
    try:
        chuoi_vao = input().split()
        if not chuoi_vao:
            return    
        cac_so = [int(x) for x in chuoi_vao]
        thong_ke = {'positives': 0, 'negatives': 0, 'zeros': 0} 
        for so in cac_so:
            if so > 0:
                thong_ke['positives'] += 1
            elif so < 0:
                thong_ke['negatives'] += 1
            else:
                thong_ke['zeros'] += 1            
        print(thong_ke)
    except (EOFError, ValueError):
        pass
dem_loai_so()
```

W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
key. 

```python
def gop_tu_dien():
    try:
        dong_1 = input().split()
        dong_2 = input().split()
        ket_qua = {}
        for i in range(0, len(dong_1), 2):
            khoa = dong_1[i]
            gia_tri = int(dong_1[i+1])
            ket_qua[khoa] = gia_tri
        for i in range(0, len(dong_2), 2):
            khoa = dong_2[i]
            gia_tri = int(dong_2[i+1])
            ket_qua[khoa] = ket_qua.get(khoa, 0) + gia_tri
        tu_dien_sap_xep = {k: ket_qua[k] for k in sorted(ket_qua)}
        print(tu_dien_sap_xep)
    except (EOFError, ValueError, IndexError):
        pass
gop_tu_dien()
```

W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple

```python
def tim_cap_tong_bang_k():
    try:
        chuoi_vao = input().split()
        if not chuoi_vao:
            return
        cac_so = [int(x) for x in chuoi_vao]
        k = int(input())  
        ket_qua = []
        n = len(cac_so)
        for i in range(n):
            for j in range(i + 1, n):
                if cac_so[i] + cac_so[j] == k:
                    ket_qua.append((cac_so[i], cac_so[j]))
        ket_qua.sort(key=lambda x: x[0])
        print(ket_qua) 
    except (EOFError, ValueError):
        pass
print(tim_cap_tong_bang_k())
```

W6A12
[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
trong số đó. 

```python
def check():
    try:
        dong_vao = input().split()
        if not dong_vao:
            return       
        cac_so = [int(x) for x in dong_vao]
        dem = {}
        for so in cac_so:
            dem[so] = dem.get(so, 0) + 1         
        so_lan_max = max(dem.values())
        cac_so_max = [so for so, tan_suat in dem.items() if tan_suat == so_lan_max]    
        print(min(cac_so_max))
    except (EOFError, ValueError):
        pass
check()
```

[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một
dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In

```python
def dao_nguoc_tu_dien():
    try:
        dong_vao = input().split()
        if not dong_vao:
            print({})
            return
        tu_dien_dao_nguoc = {}
        for i in range(0, len(dong_vao), 2):
            khoa = dong_vao[i]
            gia_tri = dong_vao[i+1]
            tu_dien_dao_nguoc[gia_tri] = khoa
        print(tu_dien_dao_nguoc)
    except (EOFError, ValueError, IndexError):
        pass
dao_nguoc_tu_dien()
```

W6A14
[Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp. 

```python
def giao_hai_danh_sach():
    try:
        dong_1 = set(map(int, input().split()))
        dong_2 = set(map(int, input().split()))  
        ket_qua = list(dong_1.intersection(dong_2))
        print(ket_qua)
    except (EOFError, ValueError):
        pass
giao_hai_danh_sach()
```

[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
def loc_tu_dien_theo_gia_tri():
    try:
        dong_vao = input().split()
        k = int(input())       
        tu_dien_moi = {}
        for i in range(0, len(dong_vao), 2):
            khoa = dong_vao[i]
            gia_tri = int(dong_vao[i+1])
            if gia_tri > k:
                tu_dien_moi[khoa] = gia_tri               
        print(tu_dien_moi)
    except (EOFError, ValueError, IndexError):
        pass
loc_tu_dien_theo_gia_tri()
```
