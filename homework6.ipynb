Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n = int(input())
m = int(input())
matrix = []
for i in range(n):
    row = []
    for j in range(m):
        val = int(input(f"nhap phan tu[{i}][{j}]: "))
        row.append(val)
    matrix.append(row)
print("\n")
for row in matrix:
    for element in row :
        print(f"{element:4d}", end = "")
    print()
        
```

W6A17 
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python

n = int(input())
ma_tran = [list(map(int, input().split())) for _ in range(n)]
cheo_chinh = []
cheo_phu = []
for i in range(n):
    cheo_chinh.append(ma_tran[i][i])            
    cheo_phu.append(ma_tran[i][n - 1 - i])     
print("Đường chéo chính:", *cheo_chinh)
print("Đường chéo phụ:", *cheo_phu)
```

trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về

W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
def vi_tri(dong_chuoi):
    danh_sach_tu = dong_chuoi.split()
    mapp = {}
    for index, tu in enumerate(danh_sach_tu):
        if tu not in mapp:
            mapp[tu] = [index]
        else:
            mapp[tu].append(index)   
    for tu in mapp:
        mapp[tu] =tuple(mapp[tu])
    return mapp
du_lieu = input()
print(vi_tri(du_lieu))
```

W6A22
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
def tim_vi_tri_lon_nhat(danh_sach):
    if not danh_sach:
        return -1
    vi_tri_max = 0
    for i in range(1, len(danh_sach)):
        if danh_sach[i] > danh_sach[vi_tri_max]:
            vi_tri_max = i       
    return vi_tri_max
danh_sach = list(map(int, input().split()))
print(tim_vi_tri_lon_nhat(danh_sach))
```

W6A23 VIết hàm xóa đi các số <x có trong mảng

```python
def xoa_phan_tu_nho_hon_x(danh_sach, x):
    i = 0
    while i < len(danh_sach):
        if danh_sach[i] < x:
            danh_sach.pop(i)
        else:
            i += 1
    return danh_sach
danh_sach = list(map(int, input().split()))
x = int(input())
print(xoa_phan_tu_nho_hon_x(danh_sach,x))
```

W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day_so_zero(danh_sach):
    ket_qua = []
    for x in danh_sach:
        if x == 0:
            ket_qua.append(0)
    for x in danh_sach:
        if x != 0:
            ket_qua.append(x)         
    return ket_qua
danh_sach =  list(map(int, input().split()))
print(day_so_zero(danh_sach))
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
def tong_cuc_dai_dia_phuong(danh_sach):
    tong = 0
    for i in range(1, len(danh_sach) - 1):
        if danh_sach[i] > danh_sach[i-1] and danh_sach[i] > danh_sach[i+1]:
            tong += danh_sach[i]
    return tong
danh_sach =  list(map(int, input().split()))
print(tong_cuc_dai_dia_phuong(danh_sach))
```

W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1

```python
def tim_vi_tri_nho_hon_x(danh_sach, x):
    vi_tri = []
    for i in range(len(danh_sach)):
        if danh_sach[i] < x:
            vi_tri.append(i)  
    if not vi_tri:
        print(-1)
    else:
        for v in vi_tri:
            print(v, end=' ')
danh_sach =  list(map(int, input().split()))
x = int(input())
print(tim_vi_tri_nho_hon_x(danh_sach,x))
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
def kiem_tra_day_so(danh_sach):
    if len(danh_sach) < 2:
        return "vo danh"
    tang = True
    giam = True
    for i in range(len(danh_sach) - 1):
        if danh_sach[i] >= danh_sach[i+1]:
            tang = False
        if danh_sach[i] <= danh_sach[i+1]:
            giam = False
    if tang:
        return "tang thuc su"
    elif giam:
        return "giam thuc su"
    else:
        return "vo danh"
danh_sach =  list(map(int, input().split()))
print(kiem_tra_day_so(danh_sach))
```
