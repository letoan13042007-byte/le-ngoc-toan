Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n = int(input())
m = int(input())
matrix = []
for i in range(n):
    row = []
    for j in range(m):
        val = int(input(f"nhap phan tu[{i}][{j}]: "))
        row.append(val)
    matrix.append(row)
print("\n")
for row in matrix:
    for element in row :
        print(f"{element:4d}", end = "")
    print()
        
```

W6A18
Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k) 

```python
from random import randrange
tongcot = 0
tonghang = 0 
m, n, k = map(int, input().split())
matrix = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(randrange(1,101))
    matrix.append(row)
for i in matrix :
    for j in i :
        print(f"{j:<2}", end = " ")
    print()
for i in range(m):
    tonghang += matrix[k - 1][i]
for j in range(n):
    tongcot += matrix[j][k - 1]
print(f"tong cot {tongcot}")
print(f"tong hang { tonghang}")
```

W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1

```python
def tim_vi_tri_dau_tien(danh_sach, k):
    for i in range(len(danh_sach)):
        if danh_sach[i] == k:
            return i  
    return -1 
danh_sach = list(map(int, input().split()))
k = int(input()) 
print(tim_vi_tri_dau_tien(danh_sach,k))
```

W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
def vi_tri(dong_chuoi):
    danh_sach_tu = dong_chuoi.split()
    mapp = {}
    for index, tu in enumerate(danh_sach_tu):
        if tu not in mapp:
            mapp[tu] = [index]
        else:
            mapp[tu].append(index)   
    for tu in mapp:
        mapp[tu] =tuple(mapp[tu])
    return mapp
du_lieu = input()
print(vi_tri(du_lieu))
```

W6A23 VIết hàm xóa đi các số <x có trong mảng

```python
def xoa_phan_tu_nho_hon_x(danh_sach, x):
    j = 0
    while j < len(danh_sach) :
        if danh_sach[j] < x :
            del danh_sach[j]
        else :
            j += 1
    print(danh_sach)      
danh_sach = list(map(int, input().split()))
x = int(input())
xoa_phan_tu_nho_hon_x(danh_sach,x)
```

W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def day(danh_sach):
    ketqua = []
    for i in range(len(danh_sach)):
        if danh_sach[i] == 0 :
            ketqua.append(0)
    for i in range(len(danh_sach)):
        if danh_sach[i] != 0 :
            ketqua.append(danh_sach[i])
    return ketqua
danh_sach = list(map(int, input().split()))
day(danh_sach)
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
def tongcucdai(danh_sach):
    tong = 0
    for i in range(len(danh_sach)):
        if danh_sach[i - 1] < danh_sach[i] < danh_sach[i + 1]:
            tong += danh_sach[i]
        else :
            continue
    return tong 
danh_sach = list(map(int, input().split()))
tongcucdai(danh_sach)
```

W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1

```python
def tim_vi_tri_nho_hon_x(danh_sach, x):
    vi_tri = []
    for i in range(len(danh_sach)):
        if danh_sach[i] < x:
            vi_tri.append(i)  
    if not vi_tri:
        print(-1)
    else:
        for v in vi_tri:
            print(v, end=' ')
danh_sach =  list(map(int, input().split()))
x = int(input())
print(tim_vi_tri_nho_hon_x(danh_sach,x))
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
def kiem_tra_day_so(danh_sach):
    if len(danh_sach) < 2:
        return "vo danh"
    tang = True
    giam = True
    for i in range(len(danh_sach) - 1):
        if danh_sach[i] >= danh_sach[i+1]:
            tang = False
        if danh_sach[i] <= danh_sach[i+1]:
            giam = False
    if tang:
        return "tang thuc su"
    elif giam:
        return "giam thuc su"
    else:
        return "vo danh"
danh_sach =  list(map(int, input().split()))
print(kiem_tra_day_so(danh_sach))
```
