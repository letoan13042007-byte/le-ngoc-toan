W7A17
Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore.
Nếu không có phần tử nào thỏa mãn thì không in gì

```python
def check(lst,n):
    mylst = []
    if not lst :
        return None
    for i in lst :
        if lst.count(i) <= n/2 :
           continue
        else :
            mylst.append(i)
    return list(set(mylst))
lst = list(map(int, input().split()))
n = int(input())
print(check(lst, n))
```

dần, lẻ giảm dần; ;

W7A19
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không
dùng bất cứ thuật toán sắp xếp nào

```python
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
result = []
i = 0
j = 0
while i < len(lst1) and j < len(lst2):
    if lst1[i] < lst2[j]:
        result.append(lst1[i])
        i += 1
    else:
        result.append(lst2[j])
        j += 1
result += lst1[i:]
result += lst2[j:]
print(result)
```

Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất.
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
names = []
n = int(input())
for i in range(n):
    name = input().strip()
    names.append(name)
names.sort()
result = max(names, key=names.count)
print(result)



```

W7A23
Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài
>=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí
tự cái.

```python
xau = input()
cac_tu = xau.split()
ket_qua = {"so_luong": 0}
for tu in cac_tu:
    if len(tu) < 2:
        continue
    dieu_kien = (tu[0].isupper(),tu[1:].islower(),tu.isalpha())  
    if all(dieu_kien):
        ket_qua["so_luong"] += 1
print(ket_qua["so_luong"])
```

̀i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
n = int(input())
def la_nguyen_to(k):
    if k < 2: return False
    for i in range(2, int(k**0.5) + 1):
        if k % i == 0: return False
    return True
ket_qua = {"bat_dau": 0, "do_dai": 0}
i = 2
while i <= n:
    if not la_nguyen_to(i):
        bat_dau_tam = i
        while i <= n and not la_nguyen_to(i):
            i += 1  
        thong_tin_day = (bat_dau_tam, i - bat_dau_tam)      
        if thong_tin_day[1] > ket_qua["do_dai"]:
            ket_qua["bat_dau"] = thong_tin_day[0]
            ket_qua["do_dai"] = thong_tin_day[1]
    else:
        i += 1
dau = ket_qua["bat_dau"]
dai = ket_qua["do_dai"]
day_hop_so = list(range(dau, dau + dai))
print(*day_hop_so)
```

W7A29
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
day_so = [int(x) for x in input().split()]
if not day_so:
    print("")
else:
    m_bd = 0
    m_len = 1
    bd_tam = 0
    len_tam = 1
    for i in range(1, len(day_so)):
        if day_so[i] > day_so[i-1]:
            len_tam += 1
        else:
            bd_tam = i
            len_tam = 1
        if len_tam > m_len:
            m_len = len_tam
            m_bd = bd_tam
    print(*day_so[m_bd : m_bd + m_len])
```

W7A30
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập
vào từ bàn phím 

```python
a = [int(x) for x in input().split()]
n = len(a)
f = [1] * n
p = [-1] * n
for i in range(n):
    for j in range(i):
        if a[j] < a[i] and f[i] < f[j] + 1:
            f[i] = f[j] + 1
            p[i] = j
id = f.index(max(f))
kq = []
while id != -1:
    kq.append(a[id])
    id = p[id]

print(*kq[::-1])
```
